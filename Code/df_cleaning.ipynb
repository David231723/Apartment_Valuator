## Using Function to explore column values 

```python
def column_explore (x):
    print("TOTAL NUMBER OF ROWS: ",len(x),"\n")
    print("- - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -\n")
    print("NUMBER OF UNIQUE VALUES:",x.nunique(),"\n")
    print("- - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - \n")
    print("NUMBER OF UNIQUE VALUES DIVIDED BY TOTAL VALUES:",round((x.nunique()/len(x)*100),2),"%\n")
    print("- - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - \n")
    print("% OF NAN VALUES IS: ",round(100*(x.isnull().sum() / len(x)),2),"%\n")
    print("- - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - \n")
    print("COLUMN DATA TYPE: ",x.dtype,"\n")
    print("- - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - \n")
    print("COLUMN UNIQUE VALUES: \n \n",x.unique(),"\n")
    print("- - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - \n")
    print("VARIABLE WEIGHT WITHIN THE TOTAL OF THE COLUMN: \n \n",round((x.value_counts() / x.value_counts().sum())*100,2),"\n")
    print("- - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - \n")
```

### Transforming 'neighborhood_id' Column

For this column we are going to apply the following transformations: 

- Split the column in two: Neighborhood & District
- Remove the information of price EUR/m2

```python
#column_explore(df['neighborhood_id'])

print(df['neighborhood_id'].unique()[0])
```

- Splitting and creating columns from the original neighborhood_id

```python
# Creating two columns from the original 'neighborhood_id' by splitting it with neighborhood & District

df[['Neighborhood','District']] = df['neighborhood_id'].str.split(' - ', n=1 ,expand=True)

# Removing unwanted price EUR/m2 per neighborhood

df['Neighborhood'] = df['Neighborhood'].str.replace(r'\(.*?\)', '').str.strip()

# Splitting both Neighborhood & District by code and name

df[['Neighborhood_code','Neighborhood_name']] = df['Neighborhood'].str.split(':', n=1 ,expand=True)
df[['District_code','District_name']] = df['District'].str.split(':', n=1 ,expand=True)

# Clean names

df['Neighborhood_code'] = df['Neighborhood_code'].apply(lambda x: re.findall('\d+', x)[0]).astype(str)
df['District_code'] = df['District_code'].apply(lambda x: re.findall('\d+', x)[0]).astype(str)

# Removing un-wanted columns

columns_del = ['Neighborhood',
               'District',
               'neighborhood_id']

df = df.drop(columns=columns_del)


#df[['Neighborhood_code','Neighborhood_name','District_code','District_name']].sample(5)
df[['District_code','District_name']]
```

- Grouping codes

```python
# create a dictionary to map codes to their corresponding values
code_map = {2677: 20, 2648: 8, 2271: 19, 2750: 19, 3607: 9, 3636: 9}

df['District_code'] = df['District_code'].astype(int)


# replace the codes with their corresponding values using the replace() method
df['District_code'] = df['District_code'].replace(code_map)
```

### n_rooms Grouping

- The number of apartments with more than 6 rooms is so little, so we will group this as '6+'

```python
df['n_rooms'] = df['n_rooms'].astype(int)

df['n_rooms'] = df['n_rooms'].apply(lambda x: 6 if x >= 6 else x)

df['n_rooms'] = df['n_rooms'].astype(str)

column_explore(df['n_rooms'])
```

### n_bathrooms Grouping

- Similar approach for n_bathrooms column values

```python

# Droping nul values. just 0.7%, so no impact at all 
df = df[pd.notnull(df['n_bathrooms'])]

# Type
df['n_bathrooms'] = df['n_bathrooms'].astype(int)

# Grouping by condition
df['n_bathrooms'] = df['n_bathrooms'].apply(lambda x: 5 if x >= 5 else x)

# Changing type
df['n_bathrooms'] = df['n_bathrooms'].astype(str)


column_explore(df['n_bathrooms'])
```

- We thing that this variable can have a significant impact on the price, so in order to have this column 'clean' we have to get rid of the 12& of the sample

```python

# Removing nan rows
#df = df[pd.notnull(df['floor'])]

# Grouping

# https://stackoverflow.com/questions/29177498/python-pandas-replace-nan-in-one-column-with-value-from-corresponding-row-of-sec

df.floor.fillna(df.house_type_id, inplace=True)


df = df[df.floor != 'Pisos']



floor_values = ['1','2','3','4','Bajo','5','6','7','8','9','Casa o chalet']

df['floor'] = df['floor'].apply(lambda x: 'Otro' if x not in floor_values else x)


column_explore(df['floor'])
```
